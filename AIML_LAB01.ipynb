# **Part 2 – Visualization of Data using matplotlib, pyplots Packages**

Generate a Data set for Health Care using the following:people,age,height,weight.

```python
people=['shiva','sunny','kirthan','prabhu','durga','rohithuuuu']
age=[19,40,25,26,30,35]
weight=[65,90,40,20,60,63]
height=[160,170,162,200,172,158]
```

```python
import matplotlib.pyplot as plt
```

**Generate Scatter plot**

```python
plt.scatter(weight,height)
plt.title("Relationship between weight and height:")
plt.ylabel("Height in meters")
plt.xlabel("Weight in kgs")
plt.show()
```

```python
plt.scatter(age,weight)
plt.title("Relationship between age and weight:")
plt.ylabel("age in yrs")
plt.xlabel("Weight in kgs")
plt.show()
```

**Generate Bar Plot**

```python
plt.figure(figsize=(12,4))
plt.title("people's weight in kgs",fontsize=16,fontstyle='italic')

#Main plot function 'bar'
plt.bar(x=people,height=weight,width=0.6,color='orange',edgecolor='k',alpha=0.6)
plt.xlabel("People",fontsize=15)

#plt.xtricks(fontsize=14,rotation=30)
#plt.ytricks(fontsize=14)
plt.ylabel("Weight (in kgs)",fontsize=5)
plt.show()
```

**Generate Histogram**

```python
import numpy as np
plt.figure(figsize=(7,4))

#Main plot function 'hist'
plt.hist(weight,color='pink',edgecolor='k',alpha=0.75,bins=4)

plt.title("Histogram of weight",fontsize=20)
plt.xlabel("weight in kgs",fontsize=15)
#plt.xtricks(fontsize=15)
#plt.ytricks(fontsize=15)
plt.show()
```
